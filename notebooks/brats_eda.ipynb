```python
import matplotlib
import matplotlib.pyplot as plt
from mis.datasets import BratsDataset
from mis.settings import BRATS_PATH

matplotlib.rcParams['figure.figsize'] = (20, 20)
plt.tight_layout(pad=0);
```

```python
data = BratsDataset(
    "train",
    BRATS_PATH,
    size=128,
    normalize=True,
    to_torch=True,
    compression_params=[80, 8]
)

len(data)
```
